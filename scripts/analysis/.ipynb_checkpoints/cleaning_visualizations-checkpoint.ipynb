# Imports

```python
# packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import missingno as msno
```

```python
# data
df = pd.read_csv("../../data/processed/DMV_dataset.csv", index_col = 0)
```

# Missing Data

```python
msno.matrix(df.sort_values(["year", "GEOID"]))
plt.savefig("../../output/plots/missing_data_year.png", dpi = 400)
```

```python
msno.matrix(df.sort_values(["GEOID", "year"]))
plt.savefig("../../output/plots/missing_data_county.png", dpi = 400)
```

# Duplicated Data

```python
sns.boxplot(y = "units", x = "state", data = df)
plt.xlabel("State")
plt.ylabel("Yearly Housing Units Construction")
plt.title("Yearly Housing Unit Construction by State")
plt.savefig("../../output/plots/misslabeled_data_state.png", dpi = 400)
```
